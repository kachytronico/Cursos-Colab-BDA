Transforma el JSON a XML:

Ejecuta el siguiente código para convertir el archivo JSON a XML.

```python
import json # json: Esta librería permite trabajar con datos en formato JSON.
import xml.etree.ElementTree as ET # xml.etree.ElementTree: Esta librería permite crear y manipular árboles XML.


# Leemos el archivo JSON cargado y lo almacenamos en la variable 'data'
with open(list(uploaded.keys())[0]) as json_file:
    data = json.load(json_file)

# Definimos una función para convertir JSON a XML
def json_to_xml(json_data, parent=None):
    # Si el JSON es un diccionario, creamos un nuevo elemento XML con la clave del diccionario
    if isinstance(json_data, dict):
        root = ET.Element(list(json_data.keys())[0])
        # Llamamos recursivamente a la función para el valor del diccionario
        json_to_xml(list(json_data.values())[0], root)
        # Si hay un elemento padre, lo agregamos al padre
        if parent is not None:
            parent.append(root)
    # Si el JSON es una lista, iteramos sobre cada elemento y llamamos recursivamente a la función
    elif isinstance(json_data, list):
        for item in json_data:
            json_to_xml(item, parent)
    # Si el JSON es un valor simple, establecemos el texto del elemento XML como el valor convertido a cadena
    else:
        parent.text = str(json_data)

# Creamos un objeto XML llamado 'xml_data'
xml_data = ET.Element('root')
# Llamamos a la función para convertir 'data' a XML y lo agregamos al elemento raíz 'xml_data'
json_to_xml(data, xml_data)

# Creamos un objeto ElementTree a partir del elemento raíz y escribimos el XML en un archivo llamado 'output.xml'
tree = ET.ElementTree(xml_data)
tree.write("output.xml")

# Imprimimos un mensaje indicando que la conversión de JSON a XML se ha completado y el archivo se ha guardado
print("La conversión de JSON a XML se ha completado y el archivo 'output.xml' se ha guardado.")
```

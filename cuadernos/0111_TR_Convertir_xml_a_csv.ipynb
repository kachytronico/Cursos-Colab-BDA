Puedes convertir este XML en un archivo CSV de la siguiente manera:

```python
import xml.etree.ElementTree as ET
import csv

# Ruta al archivo XML
xml_file_path = 'colores.xml'

# Parsear el archivo XML
tree = ET.parse(xml_file_path)
root = tree.getroot()

# Lista para almacenar los datos
data = []

# Iterar sobre los elementos del XML y extraer los datos
for color in root.findall('./arrayColores'):
    nombre = color.find('nombreColor').text
    hexadec = color.find('valorHexadec').text
    data.append([nombre, hexadec])

# Ruta para guardar el archivo CSV
csv_file_path = 'colores.csv'

# Escribir los datos en el archivo CSV
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    # Escribir el encabezado
    writer.writerow(['nombreColor', 'valorHexadec'])
    # Escribir los datos
    writer.writerows(data)

print(f'El archivo CSV ha sido creado con éxito: {csv_file_path}')
```

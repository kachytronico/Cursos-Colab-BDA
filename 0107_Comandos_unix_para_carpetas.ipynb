Tutorial de Comandos Unix en Google Colab
1. Crear, Editar y Borrar un Archivo de Texto:
Para crear un archivo de texto, utiliza el comando touch y comprueba que se ha creado usando ls:

```python
!touch archivo.txt
!ls
```

Para añadir contenido al archivo, utiliza echo con >> para añadir al final:



```python
!echo "Hola, mundo!" >> archivo.txt
!echo "¡Hola de nuevo!" >> archivo.txt
!cat archivo.txt
```

Para crear una copia del archivo antes de borrarlo y comprobar que se ha copiado, utiliza cp:



```python
!cp archivo.txt archivo_copia.txt
!ls
```

Luego, puedes borrar el archivo original:



```python
!rm archivo.txt
!ls
```

Y para comprobar que la copia del archivo todavía existe:



```python
!ls
!cat archivo_copia.txt
```

2. Crear y Borrar Carpetas:
Para crear una carpeta, utiliza el comando mkdir y comprueba que se ha creado usando ls:

```python
!mkdir carpeta
!ls
```

Para borrar una carpeta y su contenido de forma recursiva, utiliza el comando rm con la opción -r:



```python
!rm -r carpeta
!ls
```

Volvemos a crear la carpeta

```python
!mkdir carpeta
!ls
```

3. Mover y Copiar Archivos entre Carpetas:
Para mover un archivo, utiliza el comando mv:

```python
!mv archivo_copia.txt carpeta/
!ls
!ls carpeta/
```

Para copiar un archivo, utiliza el comando cp:



```python
!cp carpeta/archivo_copia.txt .
!ls
```

4. Mostrar Contenidos de Carpetas con ls:
Puedes utilizar los mismos comandos ls mencionados anteriormente para mostrar el contenido de las carpetas y verificar que los archivos se han movido o copiado correctamente.

```python
!ls -l
```

Puedes usar los parámetros típicos para mejorar la visualización de los directorios

```python
!ls -s
```

```python
!ls -ls
```

5. Ver Contenidos de Archivos de Texto:
Para ver el contenido de un archivo de texto, utiliza el comando cat:

```python
!cat archivo_copia.txt
```
